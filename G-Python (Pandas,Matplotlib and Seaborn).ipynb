```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(1,20,10)
y = x**2
```

```python
plt.plot(x,y, marker ='s', linestyle='--')
```

```python
plt.bar(x,y, color='Red')
```

```python
plt.hist(y, bins= 20)
```

```python
plt.plot(x, label="x", marker="s")
plt.plot(y, label = "y", marker = 'D')
plt.legend(loc='best')
```
